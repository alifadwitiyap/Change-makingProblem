import library

```python
import numpy as np
import matplotlib.pyplot as plt 
import time
```

Greedy method

```python
def greedyMethod(n,kembalian):
  kembalian=np.sort(kembalian)[::-1]
  count=0
  for uang in kembalian:
    while (n>=uang):
      n= n-uang
      count+=1
    
  return count
```

Recursive method

```python
def recursiveMethod(n,kembalian):
    min_uang = n
    if n in kembalian:
        return 1
    else:
        cl = []
        for uang in kembalian:
            if uang < n:
                temp = recursiveMethod(n - uang,kembalian)
                num_uang = 1 + temp
                if num_uang < min_uang:
                    min_uang = num_uang
    return min_uang
```

Dynamic Programming method

```python
def dynamicMethod(n,kembalian):
  kembalian=np.sort(kembalian)
  matrix=np.array([0],dtype=np.int)
  d=np.array([],dtype=np.int)
  

  for i in range(1,n+1):
      d=np.array([])
      for uang in kembalian:
        if i>=uang:
          d=np.append(d,matrix[i-uang])
  
      matrix=np.append(matrix,np.amin(d)+1)

  return matrix[matrix.size-1].astype(int)
```

main

```python

bench=[8,10,12,15,18,21,25,30,33,35]
kembalian=np.array([1,3,5],dtype=np.int)
arrDynamicTime=np.array([])
arrRekrusifTime=np.array([])
arrGreedyTime=np.array([])

for n in bench:

  start_time = time.time()
  minDynamic=dynamicMethod(n,kembalian)
  waktuDynamic = time.time() - start_time
  arrDynamicTime=np.append(arrDynamicTime,waktuDynamic)

  start_time = time.time()
  minRekursif=recursiveMethod(n,kembalian)
  waktuRekrusif = time.time() - start_time
  arrRekrusifTime=np.append(arrRekrusifTime,waktuRekrusif)

  start_time = time.time()
  minGreedy=greedyMethod(n,kembalian)
  waktuGreedy = time.time() - start_time
  arrGreedyTime=np.append(arrGreedyTime,waktuGreedy)

  print(f'n = {n}')
  print(f'min dynamic method= {minDynamic}, min rekrusif method= {minRekursif}, min greedy ')
  print(f'waktu dynamic method= {waktuDynamic}, waktu rekrusif method= {waktuRekrusif}, waktu greedy method {waktuGreedy} (second) \n')
```

```python
fig, graph = plt.subplots()
plt.xlim([10,35])
plt.ylim([0,0.01])
graph.set_title("dynamic vs rekrusif method at coin problem")
graph.set_ylabel("time complexity")
graph.set_xlabel("Input Size")


graph.plot(bench,arrDynamicTime,label ='dynamic method time')
graph.plot(bench,arrRekrusifTime, label ='rekrusif method time')
graph.plot(bench,arrGreedyTime, label ='greedy method time')
leg = graph.legend();
```
